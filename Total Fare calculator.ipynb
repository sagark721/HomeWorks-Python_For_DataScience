Problem Statement :

Return the 'Total Payable Amount' based on the inputs such as, Group size, Individual age, and the number of individuals with Birthday on visit day.

#RULES :

#1.

AGE(Years)  : FARE |

Up to 5     : No ticket |
5-15        : 500 |
15-60       : 1000 |
60-85       : 500 |      
ABOVE 85    : No ticket 

 

#2. If Birthday, 50% discount on individual ticket.


#3. GROUP SIZE : DISCOUNT(on Total amount) |

15-20      : 15% |
20-30      : 25% |
30+        : 35% |

```python
data=[]
grp=int(input("Enter groupsize"))
for num in range(grp):
    age=int(input("Enter age: "))
    isbday=input("is B'day, Y/N: ")
    data.append((age,isbday))
    
belowfive=[]
belowfivebd=[]
fivetofifteen=[]
fivetofifteenbd=[]
sixteentosixty=[]
sixteentosixtybd=[]
sixtytoeightyfive=[]
sixtytoeightyfivebd=[]
aboveeightyfive=[]
aboveeightyfivebd=[]
for i in data:
    if i[0]<5:
        if i[1].lower()=='n':
            belowfive.append(i[0])
        else:
            belowfivebd.append(i[0])
            
    if 5<i[0]<15:
        if i[1].lower()=='n':
            fivetofifteen.append(i[0])
        else:
            fivetofifteenbd.append(i[0])
        
    if 16<i[0]<60:
        if i[1].lower()=='n':
            sixteentosixty.append(i[0])
        else:
            sixteentosixtybd.append(i[0])
    if 60<i[0]<85:
        if i[1].lower()=='n':
            sixtytoeightyfive.append(i[0])
        else:
            sixtytoeightyfivebd.append(i[0])
    if i[0]>85:
        if i[1].lower()=='n':
            aboveeightyfive.append(i[0])
        else:
            aboveeightyfivebd.append(i[0])
        
f399=(len(fivetofifteen)*500) + 0.50*500*(len(fivetofifteenbd)) + (len(sixtytoeightyfive)*500) + (0.50*500*(len(sixtytoeightyfivebd)))
f799=(len(sixteentosixty)*1000) + 0.50*1000*(len(sixteentosixtybd))

if grp<15:
    totalfare=f399+f799
if 15<grp<20:
    totalfare=0.85*(f399+f799)
if 20<grp<30:
    totalfare=0.75*(f399+f799)
if grp>30:
    totalfare=0.65*(f399+f799)
    
print("Total Fare: ",totalfare)
```
