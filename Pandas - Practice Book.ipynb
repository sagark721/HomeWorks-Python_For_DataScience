```python
def quality(total_bill,tip):
    if tip/total_bill>0.25:
        return'Generous'
    else:
        return 'Other'
```

```python
df['Quality']=df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis=1)
```

```python
df.head()
```

```python
df['Quality_Fast']=np.vectorize(quality)(df['total_bill'],df['tip'])
```

```python
df.head()
```

```python
%timeit df['Quality']=df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis=1)
```

```python
%timeit df['Quality_Fast']=np.vectorize(quality)(df['total_bill'],df['tip'])
```

```python
import timeit
```

```python
setup='''
import numpy as np
import pandas as pd

df=pd.read_csv("tips.csv")

def quality(total_bill,tip):
    if tip/total_bill>0.25:
        return "Generous"
    else:
        return "Other"
'''
```

```python
statement_one_lambda='''
df['Quality']=df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis=1)

'''

statement_two_vectorize='''
df['Quality_Fast']=np.vectorize(quality)(df['total_bill'],df['tip'])

'''
```

```python
timeit.timeit(setup=setup,stmt=statement_one_lambda,number=1000)
```

```python
timeit.timeit(setup=setup,stmt=statement_two_vectorize,number=1000)
```

```python
df.pivot(index='Company',columns='Product',values='Licenses')
```

```python
licenses
```

```python
pd.pivot(data=licenses,index='Company',columns='Product',values='Licenses')
```

```python
pivoteddf=df.groupby(['Product','Company']).sum()['Licenses']
```

```python
pivoteddf
```

```python
pd.pivot_table(data=df,index='Company',aggfunc='sum')
```

```python
df.groupby('Company').sum()
```

```python
pd.pivot_table(data=df,index='Company',aggfunc='sum',values=['Licenses','Sale Price'])
```

```python
df
```

```python
pd.pivot_table(data=df,index=['Account Manager','Contact'],values=['Sale Price'],aggfunc='sum')
```

```python
pd.pivot_table(data=df,index=['Account Manager','Contact'],columns=['Product'],fill_value=0,values=['Sale Price'],aggfunc=[np.sum])
```

```python
pd.pivot_table(data=df,index=['Account Manager','Contact','Product'],fill_value=0,values=['Sale Price'],aggfunc=[np.sum])
```

```python
pd.pivot_table(data=df,margins=True,index=['Account Manager','Contact','Product'],fill_value=0,values=['Sale Price'],aggfunc=[np.sum])
```
